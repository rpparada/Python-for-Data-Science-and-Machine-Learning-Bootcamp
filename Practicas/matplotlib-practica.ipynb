```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional
```

```python
plt.plot(x,y)
plt.xlabel('test x')
plt.ylabel('test y')
plt.title('test plt')
#lt.plot(x,y,'r-')
#plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('test x')
axes.set_ylabel('test y')
axes.set_title('test title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('hola1')

axes2.plot(y,x)
axes2.set_title('hola2')
```

```python
#fig,axes = plt.subplots()
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
#plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('test 1')
axes[1].plot(y,x)
axes[1].set_title('test 1')
plt.tight_layout()
```

```python
axes
```

```python
# Figure Size and DPI
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(5,3))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,3))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
